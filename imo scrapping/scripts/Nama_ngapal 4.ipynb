```python
import os
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
filename = 'tanker.csv'
directory ='C:/Users/abinp/Documents/ITS/side projects/data scrapping/phase 1/outputs/'
csv_write = '{}{}'.format(directory,filename)
```

```python
%ls
link =''
vessel_names = []
vessel_type = 6 #401 bulk carrier , 402 general cargo , 403 container , 6 all tanker
minDW =0
maxDW =1500
minYear = '1985'
```

```python
def changelink(min,max):
    global minDW , maxDW , link
    minDW = min + 1500
    maxDW = max + 1500
    link = "https://www.vesselfinder.com/vessels?type={}&minDW={}&maxDW={}&minYear={}".format(vessel_type,minDW,maxDW,minYear)   
changelink(minDW,maxDW)
print(link)
```

```python
driver = webdriver.Chrome(executable_path=r"chromedriver")
driver.get(link)
```

```python
while True:
    names = driver.find_elements(By.CLASS_NAME, "slna")
    for name in names:
        vessel_names.append(name.text)
        vessel_names.append(minDW)
        vessel_names.append(maxDW)
    try:
        next_page = driver.find_element(By.CLASS_NAME, "pagination-next")
        if not next_page.is_enabled():
            print("Out of pages")
            break
        next_page.click()
    except Exception as e:
        print("continuing", e)
        changelink(minDW,maxDW)
        driver = webdriver.Chrome(executable_path=r"chromedriver")
        driver.get(link)
        continue
    if maxDW > 150000:
        break
```

```python
# assume the list of names is stored in a variable called "vessel_names"

# open a new CSV file for writing
with open(csv_write, 'w', newline='') as csvfile:
    # create a CSV writer object
    writer = csv.writer(csvfile)
    
    # write the header row (optional)
    writer.writerow(['Name', 'minDW', 'maxDW'])

    # write each name and its corresponding minDW and maxDW values to a new row in the CSV file
    for i in range(0, len(vessel_names), 3):
        name = vessel_names[i]
        minDW = vessel_names[i+1]
        maxDW = vessel_names[i+2]
        writer.writerow([name, minDW, maxDW])
```
